```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
import graphviz


pddata = pd.read_csv('agaricus-lepiota.data')

print(pddata.head())

le = preprocessing.LabelEncoder()

for col in pddata.columns.values:
        uniq_label=list(pddata[col].unique())
        #print(uniq_label)
        le_fitted = le.fit(uniq_label)
        pddata[col] = le.transform(list(pddata[col].values))

print(pddata.head())
print(pddata.shape)
```

```python
df1=pd.DataFrame(index=pddata.index)
df1=df1.join(pddata)
```

```python
df1.head()
```

```python
pddata = df1
Y_train = pddata.values[0:7000,0] # col 0 is the class label
X_train = pddata.values[0:7000,1:22] # the remaining columns are data
Y_test = pddata.values[7001:, 0]
X_test = pddata.values[7001:, 1:22]

print(X_train[0:5])
print(Y_train[0:5])
```

```python
clf=DecisionTreeClassifier() # define a new classifier clf


#clf=GaussianNB()

clf=clf.fit(X_train, Y_train) # perform training

#dot_data = tree.export_graphviz(clf, out_file="mush.png")
#graph = graphviz.Source(dot_data)
#graph.render("mushroom")

Y_predict = clf.predict(X_test)

print(accuracy_score(Y_test, Y_predict)*100)
```
